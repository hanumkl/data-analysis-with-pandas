# Tidy Data

### Melting Data

```python
import pandas as pd
df_ebola = pd.read_csv("ebola.csv")
```

```python
df_ebola.head()
```

```python
ebola_melt = pd.melt(frame=df_ebola, id_vars='Date',
       value_vars=['Cases_Guinea', 'Cases Liberia'],
       var_name='Cases', value_name='Values')
ebola_melt.head()
```

### Pivot: un-melting data

```python
ebola_tidy = ebola_melt.pivot(index='Date',
                             columns='Cases',
                             values='Values')
ebola_tidy.head()
```

```python
# ketika ada nilai duplikat pada index yg sama, 
# bisa digunakan aggregate function

import numpy as np
ebola2_tidy = ebola_melt.pivot_table(values='Values',
                                     index='Date',
                                     columns='Cases',
                                     aggfunc=np.mean)
ebola2_tidy.head()
```
